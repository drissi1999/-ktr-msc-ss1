Function to Scale Bounding Boxes

```python
def scale_bbox(box, original_size, target_size=(1000, 1000)):
    x_scale = target_size[0] / original_size[0]
    y_scale = target_size[1] / original_size[1]
    return [int(box[0] * x_scale), int(box[1] * y_scale), int(box[2] * x_scale), int(box[3] * y_scale)]
```

Function to Detect Currency in Text

```python
def detect_currency(text):
    currency_symbols = {
        '$': 'USD',
        '€': 'EUR',
        '£': 'GBP',
        '¥': 'JPY',
        '₹': 'INR',
        '₩': 'KRW',
    }
    for symbol, currency in currency_symbols.items():
        if symbol in text:
            return currency
    return None
```

Function to Preprocess Data for Model Training

```python
def preprocess_data(examples):
    images = []
    words = []
    boxes = []
    labels = []
    original_size = (224, 224)
    currency_converter = CurrencyRates()

    for image, gt in zip(examples['image'], examples['ground_truth']):
        img = image.convert("RGB").resize(original_size)
        images.append(img)
        gt = json.loads(gt)
        batch_words = []
        batch_boxes = []
        batch_labels = []

        for item in gt['valid_line']:
            for w in item['words']:
                text = w['text']
                quad = w['quad']
                bbox = scale_bbox([quad['x1'], quad['y1'], quad['x3'], quad['y3']], original_size)
                bbox = [min(max(0, coord), 1000) for coord in bbox]
                batch_words.append(text)
                batch_boxes.append(bbox)
                if item['category'] == 'total.total_price':
                    try:
                        total_amount_match = re.findall(r"\d+\.\d{2}", text)
                        if total_amount_match:
                            total_amount = float(total_amount_match[0])
                            detected_currency = detect_currency(text)
                            if detected_currency and detected_currency != 'USD':
                                total_amount = currency_converter.convert(detected_currency, 'USD', total_amount)
                            text = f"{total_amount:.2f} USD"
                    except Exception as e:
                        print(f"Error processing text: {e}")
                    batch_labels.append(label_map["total.total_price"])
                else:
                    batch_labels.append(label_map["other"])

        words.append(batch_words)
        boxes.append(batch_boxes)
        labels.append(batch_labels)

    encoding = processor(images, words, boxes=boxes, word_labels=labels, truncation=True, padding="max_length", max_length=512, return_tensors="pt")
    return encoding
```

Split Dataset into Train and Test Sets

```python
train_size = 0.8
train_indices = np.random.choice(len(encoded_small_dataset), int(len(encoded_small_dataset) * train_size), replace=False)
test_indices = np.setdiff1d(np.arange(len(encoded_small_dataset)), train_indices)

train_dataset = encoded_small_dataset.select(train_indices)
test_dataset = encoded_small_dataset.select(test_indices)

print(f"Train dataset size: {len(train_dataset)}")
print(f"Test dataset size: {len(test_dataset)}")
```

Load Pre-trained Model for Token Classification

```python
model = LayoutLMv3ForTokenClassification.from_pretrained("microsoft/layoutlmv3-base", num_labels=len(label_map))
```

Define the custom data collator for Token Classification

```python
class CustomDataCollator:
    def __call__(self, features):
        batch = {
            "input_ids": torch.stack([torch.tensor(f["input_ids"], dtype=torch.long) for f in features]),
            "attention_mask": torch.stack([torch.tensor(f["attention_mask"], dtype=torch.long) for f in features]),
            "bbox": torch.stack([torch.tensor(f["bbox"], dtype=torch.long) for f in features]),  # Changed to long
            "labels": torch.stack([torch.tensor(f["labels"], dtype=torch.long) for f in features]),
            "pixel_values": torch.stack([torch.tensor(f["pixel_values"], dtype=torch.float32) for f in features])
        }
        return batch

data_collator = CustomDataCollator()
```

Training Arguments for Fine-Tuning LayoutLMv3

```python
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    use_cpu=True
)
```

Initializing Trainer for LayoutLMv3

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    data_collator=data_collator,
    tokenizer=processor.tokenizer
)
```

```python
metric_accuracy = load_metric("accuracy", trust_remote_code=True)
metric_f1 = load_metric("f1", trust_remote_code=True)


def compute_metrics(p):
    predictions, labels = p
    predictions = np.argmax(predictions, axis=2)

    true_predictions = [
        [pred for pred, label in zip(prediction, label) if label != -100]
        for prediction, label in zip(predictions, labels)
    ]
    true_labels = [
        [label for pred, label in zip(prediction, label) if label != -100]
        for prediction, label in zip(predictions, labels)
    ]

    all_true_labels = [item for sublist in true_labels for item in sublist]
    all_true_predictions = [item for sublist in true_predictions for item in sublist]
    accuracy = metric_accuracy.compute(predictions=all_true_predictions, references=all_true_labels)
    f1 = metric_f1.compute(predictions=all_true_predictions, references=all_true_labels, average='weighted')


    return {
        "accuracy": accuracy["accuracy"],
        "f1": f1["f1"]
    }
```

```python
import torch
from transformers import LayoutLMv3ForTokenClassification, LayoutLMv3Processor
from PIL import Image

# Load your fine-tuned model and processor
model_path = "/content/drive/My Drive/fine_tuned_model"
model = LayoutLMv3ForTokenClassification.from_pretrained(model_path)
processor = LayoutLMv3Processor.from_pretrained(model_path, apply_ocr=True)




def predict_total_price(image_path, model, processor):

    image = Image.open(image_path).convert("RGB")
    encoding = processor(images=image, return_tensors="pt")
    input_ids = encoding['input_ids'][0].tolist()
    tokens = processor.tokenizer.convert_ids_to_tokens(input_ids)

    with torch.no_grad():
        outputs = model(**encoding)

    logits = outputs.logits
    predictions = torch.argmax(logits, dim=-1).squeeze().tolist()

    print(f"Predictions: {predictions}")

    total_price_text = ""
    found_total = False
    numeric_part = False

    for i, (token, pred) in enumerate(zip(tokens, predictions)):
        if "Total" in token and pred == 1:
            found_total = True
            total_price_text = ""
        elif found_total and (token.replace('Ġ', '').replace(',', '').replace('.', '').isdigit() or token in [',', '.']):
            total_price_text += token.replace('Ġ', '').strip()
            numeric_part = True
        elif found_total and numeric_part and not token.replace('Ġ', '').replace(',', '').replace('.', '').isdigit():
            break

    print(f"Total price text: {total_price_text.strip()}")

    if total_price_text:
        try:
            total_price = float(total_price_text.replace(',', '.'))
            return total_price
        except ValueError:
            return "Total price could not be converted to a number"
    else:
        return "Total price not found"

# Run the prediction on the new image
image_path = "/content/1012-receipt.jpg"

total_price = predict_total_price(image_path, model, processor)
print(f"Total Price: {total_price}")
```
